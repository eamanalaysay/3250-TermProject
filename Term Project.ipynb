## Account Balance

```python
# Account balance  mean is 2.577
print(credit_card_data['Account Balance'].median())
# 2 is the most common so the account balance for most of the people in the dataset is between 0 (inclusive) and 200 (exclusive)

credit_card_data[credit_card_data['Account Balance'] == 4]['Account Balance'].count()
# This means that 394 people in the dataset do not have a checking account


# My guess is that people with low or no checking accounts will have higher probability of being rejected for a credit card. 
# Since the company would have a higher risk of people not being able to pay ot back
```

```python
acct_plot = credit_card_data.pivot_table(index='Account Balance', values='Creditability')
acct_grp = credit_card_data.groupby(['Account Balance'])[['Creditability']]
acct_grp.mean()
acct_plot.plot.bar()
plt.ylabel('rate')
plt.title('Account Balance and Creditability')
plt.show()
```
